## Agregar datos por usuario

[Ahora que los datos están limpios, agrega los datos por usuario y por periodo para que solo haya un registro por usuario y por periodo. Esto facilitará mucho el análisis posterior.]

```python
# Calcula el número de llamadas hechas por cada usuario al mes. Guarda el resultado.


calls['call_date'] = pd.to_datetime(calls['call_date'])

# Crear una nueva columna 'month' que contenga el mes y el año de la llamada
calls['month'] = calls['call_date'].dt.to_period('M')

# Agrupar por 'user_id' y 'month', y contar el número de llamadas
llamadas_por_usuario = calls.groupby(['user_id', 'month']).size().reset_index(name='num_calls')

# Guardar el resultado en un archivo CSV
llamadas_por_usuario.to_csv('./llamadas_por_usuario.csv', index=False)

llamadas_por_usuario.head()
```

```python
# Calcula la cantidad de minutos usados por cada usuario al mes. Guarda el resultado.

import pandas as pd

calls['call_date'] = pd.to_datetime(calls['call_date'])

# Crear una nueva columna 'month' que contenga el mes y el año de la llamada
calls['month'] = calls['call_date'].dt.to_period('M')

# Agrupar por 'user_id' y 'month', y sumar la duración de las llamadas
minutos_por_usuario = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index(name='total_minutes')

# Guardar el resultado en un archivo CSV
minutos_por_usuario.to_csv('./minutos_por_usuario.csv', index=False)

minutos_por_usuario.head()
```

```python
# Calcula el número de mensajes enviados por cada usuario al mes. Guarda el resultado.


messages['message_date'] = pd.to_datetime(messages['message_date'])

# Crear una nueva columna 'month' que contenga el mes y el año del mensaje
messages['month'] = messages['message_date'].dt.to_period('M')

# Agrupar por 'user_id' y 'month', y contar los mensajes enviados
mensajes_por_usuario = messages.groupby(['user_id', 'month']).size().reset_index(name='num_messages')

# Guardar el resultado en un archivo CSV
mensajes_por_usuario.to_csv('./mensajes_por_usuario.csv', index=False)

mensajes_por_usuario.head()
```

```python
# Calcula el volumen del tráfico de Internet usado por cada usuario al mes. Guarda el resultado.


internet['session_date'] = pd.to_datetime(internet['session_date'])

# Crear una nueva columna 'month' que contenga el mes y el año del uso del Internet
internet['month'] = internet['session_date'].dt.to_period('M')

# Agrupar por 'user_id' y 'month', y sumar el volumen de tráfico de Internet
trafico_por_usuario = internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index(name='total_traffic')

# Guardar el resultado en un archivo CSV
trafico_por_usuario.to_csv('./trafico_por_usuario.csv', index=False)

trafico_por_usuario.head()
```

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month

# Fusionamos el número de llamadas con los minutos por usuario
df_merged_calls_minutes = pd.merge(llamadas_por_usuario, minutos_por_usuario, on=['user_id', 'month'], how='outer')

# Fusionamos con el número de mensajes por usuario
df_merged_calls_minutes_messages = pd.merge(df_merged_calls_minutes, mensajes_por_usuario, on=['user_id', 'month'], how='outer')

# Fusionamos con el tráfico de Internet por usuario
df_merged_final = pd.merge(df_merged_calls_minutes_messages, trafico_por_usuario, on=['user_id', 'month'], how='outer')

# Verificar el resultado
print(df_merged_final.head())

# Guardar el resultado final en un archivo CSV
df_merged_final.to_csv('./datos_fusionados.csv', index=False)
```

```python
# Añade la información de la tarifa

# Fusionamos los dataframes 'df_merged_final' y 'plans' a partir de la columna 'user_id'
df_final_with_plans = pd.merge(df_merged_final, 
                               df_users_with_plans[['plan_name', 'minutes_included', 'messages_included', 'mb_per_month_included', 
                                      'usd_per_minute', 'usd_per_message', 'usd_per_gb', 'usd_monthly_pay','user_id']], 
                               on='user_id', 
                               how='left')

# Ahora fusionamos df_final_with_plans con el dataframe 'users' para agregar la columna 'city'
df_final_with_plans = pd.merge(df_final_with_plans, 
                               users[['user_id', 'city']], 
                               on='user_id', 
                               how='left')


# Verificamos que la fusión fue exitosa
print(df_final_with_plans.head())


df_final_with_plans.to_csv('./df_final_with_plans.csv', index=False)
```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
# Calcula el ingreso mensual para cada usuario

# Definir los límites de uso gratuito y las tarifas por excedentes

# Para calcular los excedentes de minutos, mensajes y datos
df_final_with_plans['exceed_minutes'] = df_final_with_plans['total_minutes'] - df_final_with_plans['minutes_included']
df_final_with_plans['exceed_messages'] = df_final_with_plans['num_messages'] - df_final_with_plans['messages_included']
df_final_with_plans['exceed_data'] = df_final_with_plans['total_traffic'] - df_final_with_plans['mb_per_month_included']

# Asegúrate de que los excedentes no sean negativos (si el usuario no excede el límite, el excedente es 0)
df_final_with_plans['exceed_minutes'] = df_final_with_plans['exceed_minutes'].apply(lambda x: max(0, x))
df_final_with_plans['exceed_messages'] = df_final_with_plans['exceed_messages'].apply(lambda x: max(0, x))
df_final_with_plans['exceed_data'] = df_final_with_plans['exceed_data'].apply(lambda x: max(0, x))

# Ahora calculamos el costo adicional por excedentes
df_final_with_plans['extra_minutes_cost'] = df_final_with_plans['exceed_minutes'] * df_final_with_plans['usd_per_minute']
df_final_with_plans['extra_messages_cost'] = df_final_with_plans['exceed_messages'] * df_final_with_plans['usd_per_message']
df_final_with_plans['extra_data_cost'] = df_final_with_plans['exceed_data'] * df_final_with_plans['usd_per_gb']

# Finalmente, calculamos el ingreso mensual total para cada usuario
df_final_with_plans['monthly_income'] = (df_final_with_plans['usd_monthly_pay'] + 
                                          df_final_with_plans['extra_minutes_cost'] + 
                                          df_final_with_plans['extra_messages_cost'] + 
                                          df_final_with_plans['extra_data_cost'])

df_final_with_plans.head()
```

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

```python
# Prueba las hipótesis

# Filtramos los datos para los planes Ultimate y Surf
ultimate_plan = df_final_with_plans[df_final_with_plans['plan_name'] == 'ultimate']['monthly_income']
surf_plan = df_final_with_plans[df_final_with_plans['plan_name'] == 'surf']['monthly_income']

# Realizamos la prueba t
t_stat, p_value = ttest_ind(ultimate_plan, surf_plan)

# Definimos el nivel de significancia
alpha = 0.05

# Mostramos el valor t y el p-valor
print(f"Estadístico t: {t_stat}")
print(f"P-valor: {p_value}")

# Comparamos el p-valor con el nivel de significancia
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Los ingresos promedio de los planes Ultimate y Surf son diferentes.")
else:
    print("No rechazamos la hipótesis nula: No hay evidencia suficiente para decir que los ingresos promedio de los planes Ultimate y Surf son diferentes.")
```
